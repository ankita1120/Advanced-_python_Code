## ***`Countplot`***

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
import pandas as pd

diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['Outcome'] = diabetes.target

_, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))


sns.countplot(x='Outcome', data=df, ax=axes[0])
axes[0].set_title("Distribution of Outcome Categories")
axes[0].set_xlabel("Outcome Bins")
axes[0].set_ylabel("Count")

sns.countplot(x='bp' , data= df , ax=axes[1])
axes[1].set_title("Distribution of Blood Pressure")
axes[1].set_xlabel("Blood Pressure")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

Common Data Visualization Challenges and their Solutions
1. Overcrowding of data:

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(6, 6), sharex=True)

# Plot data on each subplot
axs[0].plot(x, y1, color='blue')
axs[0].set_title('Sine Wave')
axs[1].plot(x, y2, color='red')
axs[1].set_title('Cosine Wave')

# Set axis labels and add space between subplots
fig.text(0.5, 0.04, 'X', ha='center')
fig.text(0.04, 0.5, 'Y', va='center', rotation='vertical')
fig.subplots_adjust(hspace=0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')

survivours = titanic['survived'].value_counts

# 1. Stacked Bar Chart of Survivors by Passenger Class
survivours_by_class = titanic.groupby(['pclass', 'survived']).size().unstack()
survivours_by_class.plot(kind='bar', stacked=True)
plt.title('Survivors by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Survivors')
plt.show()

# 2. Pie Chart of Passengers by Gender
survivours_by_class = titanic['sex'].value_counts()
survivours_by_class.plot(kind='pie' , autopct='%1.1f%%')
plt.title('Passengers by Gender')
plt.ylabel('')
plt.show()

# 3. Histogram of Passenger Ages
titanic['age'].plot(kind='hist')
plt.title('Histogram of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 4. Scatter Plot of Passenger Age and Fare Paid

titanic.plot(kind='scatter' , x='age' , y='fare')
plt.title('Scatter Plot of Passenger Age and Fare Paid')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

# 5. Bar Chart of Survivors by Age Range
titanic = sns.load_dataset('titanic')
titanic['age_range'] = pd.cut(titanic['age'], bins=[0, 18, 30, 50, 100], labels=['0-18', '18-30', '30-50', '50+'])

survivors_by_age = titanic.groupby(['age_range', 'survived']).size().unstack()
survivors_by_age.plot(kind='bar', stacked=True)
plt.title('Survivors by Age Range')
plt.xlabel('Age Range')
plt.ylabel('Number of Survivors')
plt.show()

```

```python
import matplotlib.pyplot as plt
import matplotlib. tri  as tri
import numpy as np
ang = 40
rad = 10
radm = 0.35
radii = np.linspace(radm ,0.95 , rad)
angles = np.linspace(0, np.pi, ang)
angles = np.repeat(angles[...,np.newaxis], rad , axis = 1)
angles [: , 1::2] += np.pi / ang

x = (radii * np.cos(angles)).flatten()
y = (radii * np.sin(angles)).flatten()
z = (np.sin(4 *radii)* np.cos(4* angles)).flatten()
triang = tri.Triangulation(x,y)
triang.set_mask(np.hypot(x[triang.triangles].mean(axis = 1),
                        y[triang.triangles].mean(axis = 1))
                < radm)
tpc = plt.tripcolor (triang, z,
                    shading ='flat')
plt.set_cmap("Reds")
plt.title('matplotlib.pyplot.set_cmap() Example')
plt.show()




```

Matplotlib.pyplot.autumn():

```python
import matplotlib.pyplot as plt
import matplotlib.tri as tri
import numpy as np

ang = 40
rad = 10
radm = 0.35
radii = np.linspace(radm, 0.95, rad)

angles = np.linspace(0, 0.5 * np.pi, ang)
angles = np.repeat(angles[..., np.newaxis],
                   rad, axis = 1)
angles[:, 1::2] += np.pi / ang

x = (radii * np.cos(angles)).flatten()
y = (radii * np.sin(angles)).flatten()
z = (np.sin(4 * radii) * np.cos(4 * angles)).flatten()

triang = tri.Triangulation(x, y)
triang.set_mask(np.hypot(x[triang.triangles].mean(axis = 1),
                         y[triang.triangles].mean(axis = 1))
                < radm)

tpc = plt.tripcolor(triang, z, shading ='flat')
plt.colorbar(tpc)
plt.autumn()

plt.title('matplotlib.pyplot.autumn() function\nExample', fontweight="bold")
plt.show()
```
